```python
def remove_unnecessary_chr(df, col_name):
    df[col_name] = df[col_name].str.replace("'", "")
    df[col_name] = df[col_name].str.replace('"', "")
    df[col_name] = df[col_name].str.replace("[", "")
    df[col_name] = df[col_name].str.replace("]", "")
    return df
```

```python
import pandas as pd
import numpy as np
```

```python
mapping = pd.read_csv('./genre_mapping.csv')
```

```python
# mapping 데이터프레임에서 genre를 key, map를 value로 하는 딕셔너리 만들기 
genre_map_dict = dict(zip(mapping['genre'], mapping['map']))
genre_map_dict = {k: v if v is not np.nan else 'None' for k, v in genre_map_dict.items()}
```

```python
df = pd.read_csv('./preprocess_raw.csv')
```

```python
genre = df[['name','genre']]
```

```python
genre = remove_unnecessary_chr(genre, 'genre')
```

```python
# 장르 대분류를 저장할 세로운 데이터 프레임 생성
genre_map_df = pd.DataFrame(columns=['name', 'major_genre', 'genre'])
```

```python
lst=[]
for i in genre['genre']:
    k = i.split(',')
    k = [genre_map_dict.get(item.strip()) for item in k] # 딕셔너리를 기준으로 맵핑 
    lst_tmp = []
    for j in k:
        k_ = j.split(',')
        for z in k_:
            if z != 'None': lst_tmp.append(z) # 장르가 None이면 리스트에 추가하지 않는다. ['Adventrue', 'None', 'RPG] 등의 예시를 방지하기 위함
    if not lst_tmp: lst_tmp.append('None') # 마지막까지 리스트가 비어있으면 장르가 없다는 의미, 따라서 장르가 없다는 뜻인 'None'를 추가 
    lst.append(list(set(lst_tmp)))
```

```python
# lst 리스트에 'Adventure', ' Adventure' 와 같이 공백 으로 인해 동일한 장르지만 다르게 인식하여 중복제거가 되지 않는 문제 발생 
# 따라서 lst 리스트의 공백을 제거 후 중복을 제거한 새로운 lst2 리스트를 생성
lst = [[item.strip() for item in sublist] for sublist in lst] # lst 공백 제거

lst_2 = []
for i in lst:
    lst_2.append(list(set(i)))
```

```python
genre_map_df['name'] = genre['name']

genre_map_df['major_genre'] = lst_2
genre_map_df['major_genre']  = genre_map_df['major_genre'] .astype(str)
genre_map_df = remove_unnecessary_chr(genre_map_df, 'major_genre')

genre_map_df['genre'] = genre['genre']
```
